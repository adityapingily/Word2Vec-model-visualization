### Part 2

# Stemming the words

```python
from nltk.stem import PorterStemmer
sentences = nltk.sent_tokenize(paragraph)
stemmer = PorterStemmer()
```

```python
# Stemming
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words]
    sentences[i] = ' '.join(words) 
```

### Part 3

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
sentences = nltk.sent_tokenize(paragraph)
lemmatizer = WordNetLemmatizer()
```

```python
# Lemmatization
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word) for word in words]
    sentences[i] = ' '.join(words)  
```

### Part 4

# Stopwords

```python
from nltk.corpus import stopwords
sentences = nltk.sent_tokenize(paragraph)
```

```python
# Removing stopwords
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [word for word in words if word not in stopwords.words('english')]
    sentences[i] = ' '.join(words)
```

### Part 7 
# Creating our Bag-Of-Words Model Part-1

```python
# Tokenize sentences
dataset = nltk.sent_tokenize(paragraph)
for i in range(len(dataset)):
    dataset[i] = dataset[i].lower()
    dataset[i] = re.sub(r'\W',' ',dataset[i])
    dataset[i] = re.sub(r'\s+',' ',dataset[i])
```

### Part 8 
# Creating our Bag-Of-Words Model Part-2

```python
# Creating word histogram
word2count = {}
for data in dataset:
    words = nltk.word_tokenize(data)
    for word in words:
        if word not in word2count.keys():
            word2count[word] = 1
        else:
            word2count[word] += 1
```

### Part 9
# Creating our Bag-Of-Words Model Part-3

```python
# Selecting best 100 features
freq_words = heapq.nlargest(100,word2count,key=word2count.get)
```

### Part 10
# Creating our Bag-Of-Words Model Part-5

```python
# Converting sentences to vectors
X = []
for data in dataset:
    vector = []
    for word in freq_words:
        if word in nltk.word_tokenize(data):
            vector.append(1)
        else:
            vector.append(0)
    X.append(vector)
        
X = np.asarray(X)
```
